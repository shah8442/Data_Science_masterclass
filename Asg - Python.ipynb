```python


''' Q1. Explain with an example each when to use a for loop and a while loop.

Ans:

    for loop >: for loop is used when we know how many times we want to execute the block of code.
It iterates over a sequence (e.g., a list, tuple, string, or range) and executes the block of code for each item in the sequence.

While loop>: while loop is used when we donâ€™t know how many times we want to execute the block of code and it continues to execute till a specified condition is true.'''

## for loop example:

# print the numbers 0 to 9
for i in range(10):
    print(i)

## while loop example:

# print the numbers 0 to 9

i = 0

while i < 10:
    print(i)
    i += 1
```

```python
## ''' Q2. Write a python program to print the sum and product of the first 10 natural numbers using for and while loop.

# Ans: 
  
## for loop : 


result = 0
product = 10

for i in range(1,11):
    result = result + i
    product = product * i
print(result)
print(product)



## while loop :

sum = 0
product = 1
i = 1

while i <= 10:
    sum += i
    product *= i
    i += 1

print("The sum of the first 10 natural numbers is:", sum)
print("The product of the first 10 natural numbers is:", product)
```

```python
'''Q3. Create a python program to compute the electricity bill for a household.
The per-unit charges in rupees are as follows: For the first 100 units, the user will be charged Rs. 4.5 per
unit, for the next 100 units, the user will be charged Rs. 6 per unit, and for the next 100 units, the user will
be charged Rs. 10 per unit, After 300 units and above the user will be charged Rs. 20 per unit.
You are required to take the units of electricity consumed in a month from the user as input.
Your program must pass this test case: when the unit of electricity consumed by the user in a month is
310, the total electricity bill should be 2250.'''

Ans:

units = int(input("Enter the units of electricity consumed: "))

if units <= 100:
    bill = units * 4.5
elif units <= 200:
    bill = 100 * 4.5 + (units - 100) * 6
elif units <= 300:
    bill = 100 * 4.5 + 100 * 6 + (units - 200) * 10
else:
    bill = 100 * 4.5 + 100 * 6 + 100 * 10 + (units - 300) * 20

print("The total electricity bill is: Rs.", bill)
```

```python
'''Q5. Write a program to filter count vowels in the below-given string.
string = "I want to become a data scientist"

Ans:'''

## Ex - 1

string = "I want to become a data scientist"
vowels = "aeiouAEIOU"
count = 0

for char in string:
    if char in vowels:
        count += 1

print("The number of vowels in the string is:", count)

```
